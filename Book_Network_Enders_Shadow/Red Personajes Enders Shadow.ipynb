<h4>1.1.- Funciones auxiliares: busqueda binaria</h4>

El modo en que implementé la búsqueda de nombres de personajes en el texto fue como sigue:
primero se almacenan en un diccionario <code>characterPositions</code> todas las posiciones del texto donde aparecen los distintos personajes. Así, por ejemplo si quisieramos obtener el arreglo con todas las posiciones donde aparece Bean, llamamos: <code>characterPositions['Bean']</code>. Posteriormente lo que se hace es comparar los arreglos de posiciones, para ver cuales se encuentran a menos de 15 palabras de distancia. Supongamos que queremos comparar los arreglos de Bean y Poke, y unicamente para propósitos de ejemplo supongamos que estos son $[1,30,50,100]$ y $[20, 75, 200]$ respectivamente.
El algoritmo que implementé trata a estos datos de la siguiente manera:


<ol>
    <li>Tomamos cada uno de los elementos del arreglo de posiciones de Poke, y usamos busqueda binaria para ver en que posición caería cada una dentro del arreglo de Bean, esto nos permite hacer la comparación unicamente con las posiciones mas cercanas de aparición en ambos arreglos, lo cual hace mucho mas eficiente el código. Así, por ejemplo la primera posición donde aparece Poke, que es 20, quedaría entre el 1 y el 30 de Bean.</li>
    <li>Ahora comparamos con estos dos elementos entre los cuales queda la posición que estamos comparando, para ver si la diferencia entre alguna de ellas es menor a 15. En este caso, se tendrá al comparar con 30, que en efecto se cumple el criterio de distancia, y lo que hacemos es incrementar en 1 el peso de ambas conexiones: tanto la de Poke a Bean como la de Bean a Poke. Esta actualización doble nos asegura que el grafo resultante será no dirigido.</li>
</ol>

El código de las funciones auxiliares de busqueda binaria, y una función que regresa el código hexadecimal de un color elegido aleatoriamente (útil al dibujar el grafo) se muestran en la siguiente celda:

```python

def binarySearch(A, x, start, end):
    
    """
    BinarySearch recibe como argumento un arreglo, y el elemento que se desea buscar  
    dentro de ese arreglo. La función arroja cual es la posición que ocuparía x en A.
    Por ejemplo si A = [1,2,3,4,5] y buscamos x = 3.5, este se ubicaría entre 3 y 4, por lo cual
    tendriamos que colocar este elemento en la posición 2 (indexando desde cero). Este 2 es precisamente
    lo que devuelve el algoritmo.
    """
    
    if(start == end):
        return start
    else:
        mid = int(np.floor((start + end)/2))
        if(x <= A[mid]):
            return binarySearch(A,x,start, mid)
        elif(x > A[mid]):
            return binarySearch(A,x,mid+1, end)
        
def getRandomColor():
    
    """
    Devuelve el código hexadecimal de un color elegido aleatoriamente
    """
    s = "#"
    r = str(hex(randrange(16)))[-1:] + str(hex(randrange(16)))[-1:]
    g = str(hex(randrange(16)))[-1:] + str(hex(randrange(16)))[-1:]
    b = str(hex(randrange(16)))[-1:] + str(hex(randrange(16)))[-1:]
    s = s+r+g+b
        
    return s
```

<h4>1.2.- Implementación del grafo</h4>


La implementación del grafo se realizó utilizando una idea similar a las listas de adyacencia, esta implementación la he visto sobre todo con lenguajes de programación orientados a objetos, y resulta bastante bien por ejemplo en aplicaciones de videojuegos donde queremos hacer que algún personaje se mueva entre nodos de un mapa etc. El rendimiento, en cualquier forma resulta mucho mejor que usar una matriz de adyacencias. 

Esencialmente lo que hacemos es definir un objeto llamado <code>Connection</code> el cual representa las aristas del grafo. Este objeto tiene 3 atributos: el nodo de inicio, el nodo final y el peso del grafo. Nuestros nodos pueden ser simplemente cadenas con el nombre del nodo, o podrían ser algún objeto que encapsule mas información. En este caso para propósitos de legibilidad definí una clase <code>CharacterNode</code> la cual guarda como info el nombre del personaje en cuestión.

Nuestro grafo, cuya clase se denominó  <code>CharacterGraph</code> se conforma principalmente de un diccionario de objetos tipo <code>Connection</code> (aristas) y un arreglo de objetos de tipo <code>CharacterNode</code> (vertices): 

<ol>
    <li><code>connections</code> contiene arreglos con las conexiones de los distintos nodos. Por ejemplo <code>connections['Bean']</code> regresa un arreglo con aquellas conexiones en las que el nodo de partida sea el personaje Bean.</li>
    <li><code>nodes</code> es un arreglo que contiene todos los nodos que conforman al grafo.</li>
</ol>

El grafo contiene diversos métodos, los cuales documentamos a detalle dentro del código. Estos incluyen el cálculo de todas las medidas relevantes que se pidieron para esta tarea. 

```python
class Connection:
    """
    Connection es la clase utilizada para definir las aristas del grafo. Consta de 3 atributos: nodo de partida, 
    nodo de llegada y peso de la conexión.
    """
    
    def __init__(self, weight, startNode, endNode):
        self.startNode = startNode
        self.endNode = endNode
        self.weight = weight
        

class CharacterNode:
    
    """
    CharacterNode define los vertices del grafo de personajes. Su atributo principal es nodeId, 
    el cual es un identificador para el nodo. La igualdad de 2 objetos de tipo CharacterNode 
    se establecerá acorde a si sus identificadores son iguales.
    """
    
    #nodeId sería el nombre del personaje que conforma el nodo
    def __init__(self, nodeId):
        self.nodeId = nodeId
        
    def __eq__(self, other):
        return self.nodeId == other.nodeId
    
class CharacterGraph:
    
    """
    Clase para el grafo de personajes, conformado principalmente por un diccionario de conexiones y un arreglo 
    de nodos. 
    
    Contiene métodos para obtener las siguientes medidas de una red:
    
    - Tamaño (número de nodos y número de conexiones) y densidad
    - Camino mas corto entre dos nodos
    - Lista de caminos mas cortos entre dos nodos
    - Longitud promedio de los caminos mas cortos (L)
    - Excentricidades de los nodos, radio y diámetro de la red
    - Centralidades de grado, de intermediación y de cercanía 
    - Coeficiente promedio de agrupamiento (C)
    - Distribución de grado del grafo
    - Método Louvain para distribuir los nodos en comunidades. 
    """
    
    def __init__(self):
        self.connections = {}
        self.numberOfCharacters = 0
        self.numberOfConnections = 0
        self.nodes = []
    
    def addNode(self, node):
        """
        Añade el nodo provisto si este aun no se encuentra en el arreglo nodes
        """
        if(not node in self.nodes):
            self.nodes.append(node)
    
    def getCharacterNames(self):
        """
        Regresa una lista con el nombre de todos los personajes que conforman al grafo
        """
        return [i.nodeId for i in self.nodes]
    
    def addConnection(self, key, connection):
        """
        Añade una arista a la lista de conexiones del personaje con identificador 'key'. 
        Si aun no existe la lista para tal personaje se crea una nueva entrada en el diccionario
        de conexiones. 
        """
        #Si la conexión no existe la creamos
        if(not(key in self.connections.keys())):
            keyEdges = []
            keyEdges.append(connection)
            self.connections[key]  = keyEdges
            self.numberOfConnections += 1
            
            if(not connection.startNode in self.nodes):
                self.nodes.append(connection.startNode)
                self.numberOfCharacters += 1
            if(not connection.endNode in self.nodes):
                self.nodes.append(connection.endNode)
                self.numberOfCharacters += 1
        else:
            self.connections[key].append(connection)
            self.numberOfConnections += 1
            if(not connection.startNode in self.nodes):
                print("not existant")
                self.nodes.append(connection.startNode)
                self.numberOfCharacters += 1
            if(not connection.endNode in self.nodes):
                self.nodes.append(connection.endNode)
                self.numberOfCharacters += 1
    
    def updateConnectionWeight(self, startNodeId, endNodeId, deltaWeight):
        """
        Aumenta el peso de la conexión entre el nodo de partida startNodeId, y el nodo de llegada endNodeId.
        La cantidad en la que se aumenta la fuerza de conexión es deltaWeight. Si aun no existe la 
        conexión entre los nodos especificados se inicializa una nueva con peso deltaWeight.
        """
        key = startNodeId
        if(not(key in self.connections.keys())):
            #Creamos una nueva conexión y la agregamos
            startNode = CharacterNode(startNodeId)
            endNode = CharacterNode(endNodeId)
            connection = Connection(deltaWeight, startNode, endNode)
            self.addConnection(startNodeId, connection)
        else:
            connections = self.connections[startNodeId]
            foundConnection = False
            for i in range(0,len(connections)):
                #Buscamos una conexión
                c = connections[i]
                if(c.endNode.nodeId == endNodeId):
                    #Actualizamos el peso
                    self.connections[startNodeId][i].weight += deltaWeight
                    foundConnection = True
            if(not foundConnection):
                #Si llegamos aquí es que no existe así que creamos una nueva
                startNode = CharacterNode(startNodeId)
                endNode = CharacterNode(endNodeId)
                connection = Connection(deltaWeight, startNode, endNode)
                self.addConnection(startNodeId,connection)
    
    def getAllConnections(self):
        """
        Regresa una lista con todas las conexiones que conforman al grafo.
        """
        connectionList = []
        for key in self.connections.keys():
            nodeConnections = self.connections[key]
            connectionList.extend(nodeConnections)
            
        return connectionList
    
    
    def getMaxWeight(self):
        """
        Regresa el peso de la conexión mas fuerte en el grafo.
        """
        allConnections = self.getAllConnections()
        maxWeight = float('-inf')
        for i in range(0,len(allConnections)):
            if(allConnections[i].weight > maxWeight):
                maxWeight = allConnections[i].weight
        
        return maxWeight
    
    #Cosas que debemos calcular para esta tarea
    
    def getNetworkSizeAndDensity(self):
        """
        Regresa el número de nodos en la red (n), el número de conexiones (m)y la densidad;
        calculada como m/(n(n-1)).
        """
        n = len(self.nodes)
        m = len(self.getAllConnections())
        density = m/(n*(n-1))
        return n,m,density
    
    
    #Obtiene el camino mas corto entre 2 nodos usando bfs
    def getShortestPath(self, startNodeId, endNodeId):
        """
        Utiliza BFS (Breadth First Search) para encontrar una de las rutas mas cortas entre un nodo de partida
        startNodeId y un nodo de llegada endNodeId. Regresa una lista que contiene la secuencia de nodos a seguir
        en el camino encontrado. 
        """
        processed = []
        queue = []
        queue.append(startNodeId)
        parents = {}
        #Mientras la cola no este vacia
        foundGoal = False
        while(len(queue) != 0):
            #Sacamos el nodo que vamos a procesar
            node = queue.pop(0)
            conns = self.connections[node]
            for i in range(0,len(conns)):
                connection = conns[i]
                toNode = connection.endNode.nodeId
                if(toNode == endNodeId):
                    parents[toNode] = node
                    foundGoal = True
                    break
                
                if(not toNode in processed and not toNode in queue):
                    parents[toNode] = node
                    queue.append(toNode)
            processed.append(node)
            if(foundGoal == True):
                break
        
        path = []
        currentNode = endNodeId
        path.append(currentNode)
        while(currentNode != startNodeId):
            nextNode = parents[currentNode]
            path.append(nextNode)
            currentNode = nextNode
        
        return path[::-1]
    
    def getShortestPaths(self, startNodeId, endNodeId):
        """
        Utiliza BFS para encontrar todos los caminos de tamaño mas corto entre un nodo de partida startNodeId y un nodo
        de llegada endNodeId. Regresa un arreglo que contiene las distintas listas de nodos que corresponden a diferentes
        caminos de tamaño mas corto.
        """
        processed = []
        queue = []
        queue.append(startNodeId)
        parents = {}
        #Mientras la cola no este vacia
        paths = []
        minDist = 0
        n = len(self.nodes)
        while(len(queue) != 0):
            #print(len(queue))
            #Sacamos el nodo que vamos a procesar
            node = queue.pop(0)
            conns = self.connections[node]
            for i in range(0,len(conns)):
                connection = conns[i]
                toNode = connection.endNode.nodeId
                if(toNode == endNodeId and not (toNode in parents.keys() and parents[toNode] == startNodeId)):
                    parents[toNode] = node
                    #Extraemos el camino encontrado
                    path = []
                    currentNode = endNodeId
                    path.append(currentNode)
                    
                    while(currentNode != startNodeId):
                        nextNode = parents[currentNode]
                        path.append(nextNode)
                        currentNode = nextNode
                    #Si es el primer camino que encontramos 
                    #entonces almacenamos su longitud como la mínima
                    if(len(paths) == 0):
                        paths.append(path[::-1])
                        minDist = len(path)
                    else:
                        if(len(path) == minDist):
                            paths.append(path[::-1])
    
                elif(not toNode in processed and not toNode in queue and toNode != startNodeId):
                    parents[toNode] = node
                    queue.append(toNode)
                
                else:
                    return paths
                    
            processed.append(node)
        
        
        return paths
    
    def getL(self):
        """
        Utiliza el método getShortestPath para obtener listas Aij de los caminos entre cada par de nodos del grafo. 
        Calcula la longitud de camino como len(Aij) - 1, y obtiene el promedio de dichas longitudes. 
        """
        n = len(self.nodes)
        L = 0
        for i in range(0,n):
            fromNode = self.nodes[i].nodeId
            for j in range(0,n):
                if(i != j):
                    toNode = self.nodes[j].nodeId
                    shortestDistance = len(self.getShortestPath(fromNode, toNode)) - 1
                    L += shortestDistance/(n*(n-1))
        
        return L
                
    
    def getExcentricity(self):
        """
        Para cada nodo del grafo calcula la máxima longitud de camino hacia cualquiera de los demás nodos (excentricidad). 
        Regresa tanto el arreglo de excentricidades como el máximo de dicho arreglo (diámetro) y el mínimo (radio).
        """
        n = len(self.nodes)
        excentricities = np.zeros(n)
        #Llenamos el arreglo de excentricidades. 
        for i in range(0,n):
            fromNode = self.nodes[i].nodeId
            maxDistance = float('-inf')
            for j in range(0,n):
                toNode = self.nodes[j].nodeId
                if(i != j):
                    distance = len(self.getShortestPath(fromNode,toNode)) - 1
                    if(distance > maxDistance):
                        maxDistance = distance
            
            excentricities[i] =  maxDistance
        
        diameter = np.max(excentricities)
        radius = np.min(excentricities)
        
        return excentricities, diameter, radius
    
    def getCentralities(self):
        """
        Calcula
        - Centralidad de grado: Proporción de los nodos unidos a un nodo dado.
        - Centralidad de intermediación: Proporción de los caminos que pasan por un nodo dado, lo cual podiamos pensar
        también como la probabilidad de que al tomar un camino corto aleatorio esta pase por dicho nodo. 
        - Centralidad de cercanía: Inverso de la longitud promedio del camino de un nodo hacia los demás.
        """
        
        #Centralidad de grado 
        print("Obteniendo centralidad de grado")
        n = len(self.nodes)
        degreeCentralities = np.zeros(n)
        i = 0
        for key in self.connections.keys():
            degreeCentralities[i] = len(self.connections[key])/(n-1)
            i += 1
        
        #Centralidad de intermediación 
        intermediationCentralities = np.zeros(n)
        allPaths = []
        print("Obteniendo centralidad de intermediación")
        for i in range(0,n):
            fromNode = self.nodes[i].nodeId
            for j in range(0,n):
                toNode = self.nodes[j].nodeId
                if(i != j):
                    paths = self.getShortestPaths(fromNode, toNode)
                    nPaths = len(paths)
                    for path in paths:
                        pathLength = len(path) - 1
                        for s in range(0,pathLength):
                            nodeName = path[s]
                            if(nodeName != fromNode and nodeName != toNode):
                                index = self.nodes.index(CharacterNode(nodeName))
                                intermediationCentralities[index] += 2/(nPaths)
        
        intermediationCentralities = (1/((n-1)*(n-2)))*intermediationCentralities
        
        #Centralidad de cercanía
        closenessCentralities = np.zeros(n)
        print("Obteniendo centralidad de cercanía")
        for i in range(0,n):
            fromNode = self.nodes[i].nodeId
            for j in range(0,n):
                toNode = self.nodes[j].nodeId
                if(i != j):
                    closenessCentralities[i] += (len(self.getShortestPath(fromNode, toNode))-1)/((n-1))
        
        closenessCentralities = 1/closenessCentralities
        
        return degreeCentralities, intermediationCentralities, closenessCentralities
    
    def getAverageClusteringCoefficient(self):
        
        """
        Obtiene el coeficiente de acumulación para cada nodo, i.e. probabilidad de que los vecinos de un nodo estén 
        conextados entre sí. Regresa tanto la lista completa de coeficientes de acumulación como el promedio de dichos 
        coeficientes.
        """
        n = len(self.nodes)
        clusteringCoefficients = np.zeros(n)
        for i in range(0,n):
            degree = 2*len(self.connections[self.nodes[i].nodeId])
            #Obtenemos el vecindario del nodo 
            nodeName = self.nodes[i].nodeId
            nodeConnections = self.connections[nodeName]
            neighbours = []
            neighbours.append(self.nodes[i])
            for connection in nodeConnections:
                if(len(self.getShortestPath(nodeName, connection.endNode.nodeId)) == 2):
                    neighbours.append(connection.endNode)
            
            #Ya que tenemos el vecindario contamos las conexiones entre ellos
            connectionCount = 0
            for j in range(0,len(neighbours)):
                    startNode = neighbours[j]
                    endNodes = [no.endNode for no in self.connections[startNode.nodeId]]
                    for s in range(0,len(neighbours)):
                        endNode = neighbours[s]
                        if(j != s):
                            if(endNode in endNodes ):
                                connectionCount += 1
            
            clusteringCoefficients[i] = connectionCount/(degree*(degree - 1)/2)
            
        meanClustering = 0
        for i in range(0,n):
            meanClustering += (1/n)*clusteringCoefficients[i]
        
        return meanClustering, clusteringCoefficients
    
    def computeDegreeDistribution(self):
        """
        Regresa un arreglo con la distribución (histograma) de grados en la red, i.e. cuantos de ellos presentan 
        grado 0, grado 1 etc. También arroja el grado máximo que se obtuvo en la red. Esto es útil para por ejemplo 
        controlar acorde al grado el tamaño de los ovalos de los nodos en un dibujo del grafo. por ejemplo si queremos 
        que el máximo tamaño de ovalo sea 5, hariamos algo como tamaño = grado * 5/grado_maximo.
        """
        
        maxDegree = float('-inf')
        for key in self.connections.keys():
            if(2*len(self.connections[key]) > maxDegree):
                maxDegree = 2*len(self.connections[key])
        
        degrees = np.zeros((maxDegree))
        for key in self.connections.keys():
            degree = 2*len(self.connections[key])
            degrees[degree-1] += 1
        
        return maxDegree,degrees
    
    
    def getQ(self, m, comms):
        """
        Obtiene el valor de modularidad en un grafo con m conexiones, y una partición dada de los nodos del grafo en
        comunidades. comms es una lista de arreglos que contienen los nodos pertenecientes a cada comunidad. 
        """
        Q = 0
        n = len(self.nodes)
        for i in range(0,n):
            node1 = self.nodes[i]
            index1 = 0
            degree1 = len(self.connections[self.nodes[i].nodeId]) - 1
            for s in range(0,len(comms)):
                if(node1 in comms[s]):
                    index1 = s
                    break
                    
            for j in range(0,n):
                node2 = self.nodes[j]
                index2 = 0
                degree2 = len(self.connections[self.nodes[j].nodeId]) - 1
                for s in range(0,len(comms)):
                    if(node2 in comms[s]):
                        index2 = s
                        break
                if(i != j):
                    if(index1 == index2):
                        Q += (1/(2*m))*(2 - ((degree1)*(degree2)/(2*m)))
        return Q

    def louvainMethod(self):
        """
        Aplica el método Louvain para particionar el grafo en comunidades. Regresa una lista de arreglos que contienen
        los nodos pertenecientes a cada comunidad. 
        """
        
        m = len(self.getAllConnections())
        #Inicialmente las comunidades son iguales a los nodos
        communityCount = len(self.nodes)
        communities = []
        for i in range(0,communityCount):
            communities.append([self.nodes[i]])
        #Obtenemos los vecinos de cada nodo
        #como aquellos nodos que se ubican a una distancia 1 del nodo
        neighbours = {}
        for i in range(0,len(self.nodes)):
            fromNode = self.nodes[i]
            for j in range(0,len(self.nodes)):
                toNode = self.nodes[j]
                if(i != j):
                    radius = len(self.getShortestPath(fromNode.nodeId, toNode.nodeId)) - 1
                    if(radius == 1):
                        if(fromNode.nodeId in neighbours.keys()):
                            neighbours[fromNode.nodeId].append(toNode)
                        else:
                            neighbours[fromNode.nodeId] = []
                            neighbours[fromNode.nodeId].append(toNode)
        #Cuando la modularidad ya no cambie mucho entonces 
        #terminamos el ciclo
        qTolerance = 0.001
        deltaQ = float('inf')
        while(deltaQ > qTolerance):
            for i in range(0,len(self.nodes)):
                node = self.nodes[i]
                closeNeighbours = neighbours[node.nodeId]
                currentIndex = 0
                deltaQ = 0
                prevQ = self.getQ(m, communities)
                for s in range(0,len(communities)):
                    if(node in communities[s]):
                        currentIndex = s
                        break
                maxDelta = float('-inf')
                maxNeighbourIndex = 0
                maxNeighbour = None
                for j in range(0,len(closeNeighbours)):
                    neighbourIndex = 0
                    neighbourNode = closeNeighbours[j]
                    for s in range(0,len(communities)):
                        if(neighbourNode in communities[s]):
                            neighbourIndex = s
                            break
                    #Hacemos una copia de las comunidades
                    communitiesCopy = [k.copy() for k in communities]
                    communitiesCopy[neighbourIndex].remove(neighbourNode)
                    communitiesCopy[currentIndex].append(neighbourNode)
                    newQ = self.getQ(m, communitiesCopy)
                    if(newQ - prevQ > maxDelta):
                        maxDelta = newQ - prevQ
                        maxNeighbourIndex = neighbourIndex
                        maxNeighbour = neighbourNode
                        
                if(maxDelta > 0):
                    deltaQ += maxDelta
                    communities[maxNeighbourIndex].remove(maxNeighbour)
                    if(len(communities[maxNeighbourIndex]) == 0):
                        communities.pop(maxNeighbourIndex)
                    communities[currentIndex].append(maxNeighbour)
        return communities
```

<h4>1.2.- El constructor de la red de personajes</h4>

Se creo una clase llamada <code>BookNetworkBuilder</code> para leer el archivo de texto del libro y construir el gráfico a partir de este. Dicha clase recibe como argumentos de inicialización dos cadenas: <code>charactersPath</code>, la cual es una ruta a un archivo .txt que contiene enlistados los personajes que conformaran los nodos de la red, y <code>bookPath</code>: la ruta al archivo .txt del libro como tal. 

La lectura de los personajes (nodos del grafo) se hace utilizando el método <code>readCharacters</code>. Para construir las conexiones se requiere de dos métodos:

<ol>
    <li><code>readBook</code>: Hace algunas acciones de preprocesamiento en las cuales se eliminan signos de puntuación, saltos de linea y espacio, terminando así con un arreglo grande que contiene como entradas todas las palabras del libro.</li>
    <li><code>buildConnections</code>: En este método en primer lugar se obtienen todas las posiciones del texto en las que se encuentre el nombre de cada personaje, guardando estas en un diccionario. Posteriormente, tal y como se explicó en la sección 1.1 se utiliza busqueda binaria para comparar los arreglos de posiciones de distintos personajes, encontrando aquellas apariciones que se ubiquen a menos de 15 palabras de distancia e incrementando la fuerza de las conexiones.</li>
</ol>

```python
class BookNetworkBuilder:
    
    """
    Clase utilizada para construir un grafo de personajes a partir de una lista de personajes en .txt ubicada en 
    charactersPath y el texto del libro en .txt ubicado en la ruta bookPath.   
    """
    def __init__(self, charactersPath, bookPath):
        self.charactersPath = charactersPath
        self.bookPath = bookPath
        self.charactersGraph = CharacterGraph()
        self.deltaWeight = 1
    
    def readCharacters(self):
        """
        Genera los nodos del grafo a partir de la lista de personajes.
        """
        characterFile = open(self.charactersPath,'r')
        for characterLine in characterFile:
            node = CharacterNode(characterLine.rstrip())
            self.charactersGraph.addNode(node)
    
    def readBook(self):
        """
        Procesa el texto del libro, removiendo puntuación, espacios y saltos de linea, y almacena 
        un arreglo con todas las palabras que contenga la obra.
        """
        #Abrimos el libro. 
        bookFile = open(self.bookPath, 'r', encoding = "utf8")
        #Hagamos una lista con todas las lineas del libro
        lineArray = list(enumerate(bookFile))
        #Cortamos las tuplas que genera esto
        lineArray = [i[1] for i in lineArray]
        #Quitamos los simbolos de salto de linea
        lineArray = [i.rstrip().translate(str.maketrans('', '', string.punctuation)) for i in lineArray]
        #Quitamos las lineas vacias
        lineArray = [i for i in lineArray if i != '']
        #Almacenemos todo en un arreglo masivo de palabras
        words = []
        for l in lineArray:
            wArr = l.split(' ')
            words.extend(wArr)
        
        self.words = words
        print(len(self.words))
            
    
    
    def buildConnections(self, maxWordSeparation):
        """
        Construye las conexiones del grafo a partir del arreglo de palabras guardado con la función readBook. 
        Para ellos se obtienen las posiciones de aparición de cada personaje en el texto y se comparan para 
        ver cuales pares de ellas se encuentran a menos de 15 palabras de distancia. La función utiliza busqueda
        binaria para la comparación, lo cual vuelve al algoritmo un tanto mas eficiente. 
        """
        #Obtengamos los nombres de los personajes.
        characterNames = self.charactersGraph.getCharacterNames()
        words = np.array(self.words)
        characterPositions = {}
        #Primero vamos a obtener las posiciones donde aparece cada personaje
        for i in range(0,len(characterNames)):
            #Obtenemos el personaje
            character = characterNames[i]
            #Vemos las posiciones donde aparece
            indexes = np.where(words == character)[0]
            #print("N.posiciones "+str(len(indexes)))
            characterPositions[character] = indexes
        
        
        #Ahora si, creamos las conexiones entre los personajes
        for i in range(0,len(characterNames)):
            #clear_output(wait = True)
            #print("Personaje "+str(i),flush = True)
            character1 = characterNames[i]
            character1Indices = characterPositions[character1]
            nIndices1 = len(character1Indices)
            for j in range(i+1, len(characterNames)):
                character2 = characterNames[j]
                character2Indices = characterPositions[character2]
                for s in range(0,len(character2Indices)):
                    pos = character2Indices[s]
                    #Busqueda binaria 
                    index = binarySearch(character1Indices,pos,0,nIndices1-1)
                    position1Plus = character1Indices[index]
                    position1Minus = position1Plus
                    if(index != 0):
                        position1Minus = character1Indices[index - 1]
                        
                    if(np.abs(pos - position1Plus)<maxWordSeparation or np.abs(pos - position1Minus)<maxWordSeparation):
                        #Reforzamos la conexión 2 dos vias entre los personajes.
                        self.charactersGraph.updateConnectionWeight(character1, character2, self.deltaWeight)
                        self.charactersGraph.updateConnectionWeight(character2, character1, self.deltaWeight)
```

<h4>1.3.- El renderizador de la red</h4>

Finalmente se implementó un renderizador para lea red, el cual encapsula todos los métodos para dibujarla.  Este recibe como argumentos de inicialización 3 parámetros, un objeto de tipo <code>BookNetworkBuilder</code> el cual ya debe contener una red construida, y tanto el largo como el ancho del mundo donde se dibujará la red (1000 x 1000 suele funcionar bastante bien).

Para un despliegue básico de la red puede usarse en primer lugar la función  <code>assignPositionsAndColors</code>. Esta asigna posiciones y colores aleatorios a los nodos, guardandolos en un arreglo. Para la distribución de los puntos se utiliza un disco de Poisson (vease <url>https://medium.com/@hemalatha.psna/implementation-of-poisson-disc-sampling-in-javascript-17665e406ce1</url>), esto nos asegura en cierta medida que no habrá un traslape en los nodos de la red, lo cual ocurriría con bastante frecuencia si usaramos una distribución uniforme para asignarles una posición.

Ya asignados el color y la posición se puede utilizar la función  <code>displayGraph</code> para mostrar la red, aquí utilicé unicamente matplotlib, y creo que queda bastante presentable el grafo. 

Si se desea colorear por comunidad podemos mandar a llamar al método: <code>drawCommunities</code>, este manda a llamar al método Louvain para encontrar la partición de los nodos en comunidades, asigna colores acorde a esta y manda a llamar al método <code>displayGraph</code>.




```python
class BookNetworkRenderer:
    """
    Clase dedicada al despliegue visual de una red de personajes, así como algunas de sus características (distribución
    de grado y comunidades).
    """
    
    
    def __init__(self, bookNetworkBuilder, worldWidth, worldHeight):
        self.bookNetwork = bookNetworkBuilder
        self.worldWidth = worldWidth
        self.worldHeight = worldHeight
        self.connectionColor = "#34d8eb"
        self.maxConnectionWidth  = 5
        self.nodeSize = 2
        self.maxMarkerSize = 100
        self.nodeAlpha = 0.5
        
    #Asignamos aleatoriamente posiciones para los nodos
    def assignPositionsAndColors(self):
        
        """
        Asigna posiciones y colores aleatorios a los nodos de la red y los guarda. La distribución de posiciones se hace 
        acorde a un disco de Poisson, de modo que no haya traslape entre los ovalos que representan a los nodos.  
        """
        n = len(self.bookNetwork.charactersGraph.nodes)
        nodeColorPositions = {}
        maxConnections = float('-inf')
        poissonDisc = poisson_disc_samples(width=self.worldWidth, height=self.worldHeight, r=self.maxMarkerSize/2)
        poissonPositions = random.sample(poissonDisc, n)
        #Obtenemos el número máximo de conexiones que hay para los nodos del 
        #grafo
        i = 0
        for key in self.bookNetwork.charactersGraph.connections.keys():
            numberOfConnections = len(self.bookNetwork.charactersGraph.connections[key])
            if(numberOfConnections > maxConnections):
                maxConnections = numberOfConnections
            
            #Obtenemos un color aleatoriamente
            color = getRandomColor()
            #Obtenemos una posición aleatoria
            randX = poissonPositions[i][0]
            randY = poissonPositions[i][1]
            nodeColorPositions[key] = [randX, randY, color]
            i += 1
        
        self.nodeColorPositions = nodeColorPositions
        self.maxConnections = maxConnections
    
    def displayGraph(self):
        """
        Despliega el grafo de personajes. Es necesario haber llamado antes a la función 
        assignPositionsAndColors.
        """
        #Obtengamos todas las conexiones
        allConnections = self.bookNetwork.charactersGraph.getAllConnections()
        n = len(allConnections)
        maxWeight = self.bookNetwork.charactersGraph.getMaxWeight()
        print(maxWeight)
        #Graficamos las conexiones
        plt.figure(figsize = (20,20))
        for i in range(0,n):
            connection = allConnections[i]
            #print(maxWeight)
            connectionWidth = connection.weight*(self.maxConnectionWidth/maxWeight)
            startNodeId = connection.startNode.nodeId
            endNodeId = connection.endNode.nodeId
            
            startNodeColorPosition = self.nodeColorPositions[startNodeId]
            endNodeColorPosition = self.nodeColorPositions[endNodeId]
            
            plt.plot([startNodeColorPosition[0],endNodeColorPosition[0]],[startNodeColorPosition[1],endNodeColorPosition[1]],color = self.connectionColor, linewidth = connectionWidth)
        
        #Graficamos los nodos
        for key in self.nodeColorPositions.keys():
            nodeColorPosition = self.nodeColorPositions[key]
            markerSize = len(self.bookNetwork.charactersGraph.connections[key])*self.maxMarkerSize/self.maxConnections
            plt.plot(nodeColorPosition[0],nodeColorPosition[1],'o', markersize = markerSize, color = nodeColorPosition[2],alpha = self.nodeAlpha)
            plt.text(nodeColorPosition[0],nodeColorPosition[1]+markerSize/2, key, ha = 'center')
    
    def plotDegreeDistribution(self):
        """
        Obtiene y grafica la distribución de grados de la red.
        """
        maxDegree, degrees = self.bookNetwork.charactersGraph.computeDegreeDistribution()
        xInterval = range(1,maxDegree+1)
        yInterval = degrees
        plt.figure(figsize = (7,5))
        plt.plot(xInterval, yInterval, color = self.connectionColor)
        plt.title("Distribución de grados")
        plt.xlabel("Grado")
        plt.ylabel("Frecuencia")
    
    def plotLogLogDegreeDistribution(self):
        """
        Realiza un ajuste lineal log log a la distribución de grado, para examinar
        si se cumple una ley de potencias, lo cual sería indicativo de que la red 
        es libre de escala. 
        """
        maxDegree, degrees = self.bookNetwork.charactersGraph.computeDegreeDistribution()
        xInterval = np.array(range(1,maxDegree+1))
        yInterval = np.array(degrees)
        logX = []
        logY = []
        n = np.size(xInterval, 0)
        for i in range(0, n):
            if(yInterval[i] != 0):
                logX.append(np.log(xInterval[i]))
                logY.append(np.log(yInterval[i]))
        xVals = np.linspace(min(logX),max(logX),1000)
        coeffs = np.polyfit(logX, logY, 1)
        yhat = np.polyval(coeffs, logX)                     
        ybar = np.sum(logY)/len(logY)    
        ssreg = np.sum((yhat-ybar)**2)   
        sstot = np.sum((logY - ybar)**2)    
        rSq = ssreg / sstot
        
        yVals = np.polyval(coeffs,xVals)
        plt.figure(figsize = (7,5))
        plt.title("Ajuste Log - Log")
        plt.plot(logX, logY, 'o', color = self.connectionColor)
        plt.plot(xVals, yVals, color = "#ffce2e")
        plt.text((min(xVals) + max(xVals))/2, 1.5, "y = "+str(np.round(coeffs[0],2)) +"x + "+str(np.round(coeffs[1],2)))
        plt.text((min(xVals) + max(xVals))/2, 1.4, "r^2 = "+str(rSq))
        plt.xlabel("log(grado)")
        plt.ylabel("log(frecuencia)")
        
    def drawCommunities(self):
        """
        Utiliza el método Louvain para particionar la red en comunidades, asignando colores acorde al grupo al 
        que pertenezca el nodo. Posteriormente manda a llamar al método displayGraph para desplegar la red. 
        """
        #Asignamos colores acorde a la comunidad a la que pertenezca el nodo
        communities = self.bookNetwork.charactersGraph.louvainMethod()
        colors = []
        while(len(colors) < len(communities)):
            newColor = getRandomColor()
            if(not newColor in colors):
                colors.append(newColor)
        
        for key in self.nodeColorPositions.keys():
            n = CharacterNode(key)
            color = None
            for s in range(0, len(communities)):
                if(n in communities[s]):
                    color = colors[s]
                    break
            
            self.nodeColorPositions[key][2] = color 
        #Desplegamos el grafo
        self.displayGraph()
        
```
